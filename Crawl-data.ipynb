### 2.1. Import thÆ° viá»‡n cáº§n thiáº¿t

````python
import csv
import time
import requests
import json
````

### 2.2. CONSTANT

````python
BASE_URL = 'https://gateway.chotot.com/v1/public/ad-listing'

LIMIT = 100
LIMIT_PAGE = 200
SLEEP_TIME = 2
REGION_V2 = 13000
CG_APARTMENT = 1010
````

### 2.3. Láº¥y danh sÃ¡ch cÄƒn há»™ chÆ°a Ä‘áº§y Ä‘á»§ thÃ´ng tin

HÃ m láº¥y thÃ´ng tin 100 record má»—i lÆ°á»£t

````python
def get_data_per_page(page):
    url = f'{BASE_URL}?region_v2={REGION_V2}&cg={CG_APARTMENT}&limit={LIMIT}&o={LIMIT * page}&st=u,h&page={page}&w=1'
    res = requests.get(url)
    time.sleep(SLEEP_TIME)
    data = res.json()
    return data['ads']
````


CÃ¡ch thá»±c hiá»‡n: Crawl láº§n lÆ°á»£t 200 page, má»—i page 100 record

````python
# get all apartments data
def get_data():
    all_apartments = []
    for i in range(LIMIT_PAGE):
        print(f'get data in page {i}')
        list_apartment = get_data_per_page(i)
        all_apartments.extend(list_apartment)

    return all_apartments
````

### 2.4. Láº¥y thÃ´ng tin tá»«ng cÄƒn há»™

Láº¥y danh sÃ¡ch `list_id` Ä‘á»ƒ truy váº¥n tá»«ng cÄƒn há»™

CÃ³ chÃ­nh xÃ¡c 19913 tin cho thuÃª nhÃ  á»Ÿ thÃ nh phá»‘ Há»“ ChÃ­ Minh, cáº§n láº¥y thÃ´ng tin tá»«ng cÄƒn há»™ má»™t

````python
def get_apartment_detail(list_id):
    url = f'{BASE_URL}/{list_id}'
    res = requests.get(url)
    data = res.json()
    return data
````

## 2.5. Filter data

````python
need_keys = ['ad_id', 'list_id', 'list_time', 'subject', 'body', 'price', 'price_string', 'size', 'rooms', 'toilets', 'area_name', 'region_name', 'ward_name', 'street_name', 'deposit', 'zero_deposit', 'longitude', 'latitude', 'owner', 'company_ad', 'apartment_type']

def filter_apartment(apartment):
    new_apartment = { key: apartment['ad'][key] if key in apartment['ad'] else '' for key in need_keys }
    if apartment['ad_params']['apartment_type']['value']:
        new_apartment['apartment_type'] = apartment['ad_params']['apartment_type']['value']
    if new_apartment['deposit'] == '':
        new_apartment['deposit'] = 0
    return new_apartment

all_apartments1 = list(filter(lambda a: 'ad' in a,all_apartments))

new_apartments = list(map(filter_apartment, all_apartments1))
len(new_apartments)
````
